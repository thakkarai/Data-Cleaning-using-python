```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"D:\DATA_SCIENCE\Titanic-Dataset.csv")
print(df.tail())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.dropna()
```

```python
df.dropna(subset=['Fare'], inplace=True)
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
print(df.describe())
```

```python
import matplotlib.pyplot as plt
df['Age'].hist(bins=10, edgecolor='red')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()
```
